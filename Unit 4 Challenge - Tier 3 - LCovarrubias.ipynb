**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:

2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

```python
#visualize the average price over the years to get an idea of which boroughs may have the highest ratios
pivot_avg_price_per_year = avg_price_per_year.pivot_table(values='avg_price', index='year', columns='Boroughs_London')

# Plot the data
pivot_avg_price_per_year.plot(kind='line', figsize=(15, 10), title='Average Price per Year for Different Boroughs')
#plt.xlabel('Year')
#plt.ylabel('Average Price')
#plt.legend(title='Boroughs')
#plt.show()
```

```python
#here i realized that the index was messing up my code because later i am calling columns that exist as an index
# which further complicated my code, I ended up adding as_index=False to make sure my code executed correctly
print(avg_price_per_year.columns)
```

```python
#this function creates the average price ratio for the years 1998-2018, i am not sure i understand the syntax
def create_price_ratio(d):
    y1998 = float(d['avg_price'][d['year']==1998])
    y2018 = float(d['avg_price'][d['year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
#sampling executing the ratio for one borough to make sure the function executes correctly and it did!
create_price_ratio(avg_price_per_year[avg_price_per_year['Boroughs_London']=='Barking & Dagenham'])
```

```python
#creates an empty dictionary to place the key value pairs of borough to average price ratio
final = {}
```

```python
#iterates over all boroughs to compute all average price ratios and inserst them into the empty dictionary
for b in avg_price_per_year['Boroughs_London'].unique():
    # Let's make our parameter to our create_price_ratio function: i.e., we subset dfg on 'London_Borough' == b. 
    borough = avg_price_per_year[avg_price_per_year['Boroughs_London'] == b]
    # Make a new entry in the final dictionary whose value's the result of calling create_price_ratio with the argument: borough
    final[b] = create_price_ratio(borough)
# We use the function and incorporate that into a new key of the dictionary 
print(final) 
```

```python
#creates a dataframe from the dictionary final
ratio_boroughs = pd.DataFrame(final)
ratio_boroughs.head()
```

```python
#the dataframe is in an undesirable format and needs to be transposed anf the index also needs to be reset
ratios_tboroughs = ratio_boroughs.transpose()
ratios_riboroughs = ratios_tboroughs.reset_index()
ratios_riboroughs.head()
```

```python
#renames the columns
ratios_riboroughs.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
ratios_riboroughs.head()
```

```python
#sort values from highest to lowest ratios, only the top 15
complete_ratios = ratios_riboroughs.sort_values(by='2018',ascending=False).head(15)
print(complete_ratios)
```

```python
#creates a bar plot from the top 15 ratios
plot_ratios = complete_ratios[['Borough','2018']].plot(kind='bar')

plot_ratios.set_xticklabels(complete_ratios.Borough)
```
